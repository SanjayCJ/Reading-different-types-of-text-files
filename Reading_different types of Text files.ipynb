```
from pdfminer.high_level import extract_text
```

```
text=extract_text('/content/drive/My Drive/Types of text document/Policies and Guidelines for Effective Classroom Management.pdf')
print(text)
```

**The code above will extract the text from each page in the PDF. If we want to limit our extraction to specific pages, we just need to pass that specification to extract_text using the page_numbers parameter.**

```
#extract the text from 3rd page of the pdf
text3=extract_text('/content/drive/My Drive/Types of text document/Policies and Guidelines for Effective Classroom Management.pdf',page_numbers=[2])
print(text3)
```
